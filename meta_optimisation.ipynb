```python
# Définition des bornes réalistes par colonne
physical_limits = {
    'T (degC)': (-50, 50),        # Température réaliste en extérieur
    'p (mbar)': (800, 1200),      # Pression atmosphérique standard
    'rh (%)': (0, 100),           # Humidité relative (0-100%)
    'wv (m/s)': (0, 100),         # Vitesse du vent (positive et < limite extrême)
    'hour': (0, 23),              # Heures de 0 à 23
    'month': (1, 12),             # Mois de 1 à 12
    'year': (2009,2017)
}

# Détection des valeurs hors bornes
outliers_physical = {}
for col, (min_val, max_val) in physical_limits.items():
    outliers = df1_hourly[(df1_hourly[col] < min_val) | (df1_hourly[col] > max_val)]
    if not outliers.empty:
        outliers_physical[col] = outliers
        print(f"Outliers physiques dans {col}:\n", outliers.head(), "\n")
```

```python
valid_values = df1_hourly[(df1_hourly['wv (m/s)'] >= physical_limits['wv (m/s)'][0]) &
               (df1_hourly['wv (m/s)'] <= physical_limits['wv (m/s)'][1])]['wv (m/s)']
median_value = valid_values.median()

df1_hourly.loc[(df1_hourly['wv (m/s)'] < physical_limits['wv (m/s)'][0]) |
               (df1_hourly['wv (m/s)'] > physical_limits['wv (m/s)'][1]),
               'wv (m/s)'] = median_value
```

```python
import numpy as np

def create_sequences(data, target_col, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        # Features : les 'n_steps' dernières valeurs (exclut la cible)
        X.append(data.iloc[i:i+n_steps].drop(columns=[target_col]))
        # Cible : la valeur suivante de la colonne target_col
        y.append(data.iloc[i+n_steps][target_col])
    return np.array(X), np.array(y)

# Paramètres
n_steps = 24  # Utiliser 24 heures passées
target_col = 'T (degC)'  # Colonne à prédire

# Création des séquences
X, y = create_sequences(df1_hourly, target_col, n_steps)
```

```python
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD, RMSprop

def build_lstm_model(input_shape, optimizer='adam', learning_rate=0.001, momentum=0.0):
    """
    Construit un modèle LSTM pour la prédiction de séries temporelles.

    Args:
        input_shape (tuple): Forme des données d'entrée (timesteps, features).
        optimizer (str): 'adam', 'sgd', ou 'rmsprop'.
        learning_rate (float): Taux d'apprentissage.
        momentum (float): Momentum (pour SGD uniquement).
    """
    model = Sequential([
        # Première couche LSTM avec retour des séquences pour la couche suivante
        LSTM(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),  # Réduit le surajustement
        # Deuxième couche LSTM
        LSTM(32),
        Dropout(0.2),
        # Couche de sortie
        Dense(1)
    ])

    # Configuration de l'optimiseur
    if optimizer == 'adam':
        opt = Adam(learning_rate=learning_rate)
    elif optimizer == 'sgd':
        opt = SGD(learning_rate=learning_rate, momentum=momentum)
    elif optimizer == 'rmsprop':
        opt = RMSprop(learning_rate=learning_rate)

    model.compile(optimizer=opt, loss='mse', metrics=['mae'])
    return model
```

```python
# Pour y_train et y_test
y_train = np.nan_to_num(y_train, nan=np.nanmean(y_train))
y_test = np.nan_to_num(y_test, nan=np.nanmean(y_test))

# Pour X_train et X_test
def clean_nan_3d(array):
    """Remplace les NaN par la moyenne de la feature correspondante"""
    for feature_idx in range(array.shape[2]):
        feature_mean = np.nanmean(array[:, :, feature_idx])
        array[:, :, feature_idx] = np.nan_to_num(array[:, :, feature_idx], nan=feature_mean)
    return array

X_train = clean_nan_3d(X_train)
X_test = clean_nan_3d(X_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.callbacks import EarlyStopping

input_shape = (X_train.shape[1], X_train.shape[2])  # (timesteps, features)

# Liste des optimiseurs à tester
optimizers = ['adam', 'sgd', 'rmsprop']
results = {}

for opt_name in optimizers:
    print(f"\n=== Test avec {opt_name.upper()} ===")

    # Configuration spécifique
    lr = 0.001
    momentum = 0.9 if opt_name == 'sgd' else 0.0

    # Construction du modèle
    model = build_lstm_model(
        input_shape=input_shape,
        optimizer=opt_name,
        learning_rate=lr,
        momentum=momentum
    )

    # Entraînement avec chronométrage et early stopping
    start_time = time.time()
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=100,  # Large valeur car early stopping interviendra
        batch_size=32,
        callbacks=[EarlyStopping(monitor='val_loss', patience=5)],
        verbose=1
    )
    training_time = time.time() - start_time

    # Évaluation
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)

    # Stockage des résultats
    results[opt_name] = {
        'model': model,
        'history': history.history,
        'training_time': training_time,
        'mae': mae,
        'predictions': y_pred.flatten()
    }
```

```python
import numpy as np

def predict_temperature(model, input_features):
    """
    Prédit la température à partir des features d'entrée.

    Args:
        model: Modèle LSTM entraîné
        input_features: Liste ou tableau des valeurs des features dans cet ordre:
            [p (mbar), rh (%), wv (m/s), hour_sin, hour_cos, month_sin, month_cos]

    Returns:
        float: Température prédite en °C
    """
    # Convertir l'entrée en tableau numpy et le remodeler
    features_array = np.array(input_features).reshape(1, -1)

    # Vérifier qu'on a bien 7 features
    if features_array.shape[1] != 7:
        raise ValueError(f"7 features attendues, {features_array.shape[1]} reçues")

    # Créer une séquence de n_steps avec les mêmes valeurs (pour exemple simple)
    # En pratique, vous devriez avoir 24 pas de temps historiques
    sequence = np.tile(features_array, (24, 1)).reshape(1, 24, 7)

    # Faire la prédiction
    prediction = model.predict(sequence)

    return prediction[0][0]

# Exemple d'utilisation avec le meilleur modèle (remplacez par votre modèle préféré)
best_model = results['sgd']['model']  # Utilisation du modèle Adam comme exemple

# Exemple de features d'entrée (remplacez par vos valeurs)
sample_features = [
    0.950400,   # p (mbar)
    0.998117,    # rh (%)
    -1.255306,   # wv (m/s)
    0.866025,    # hour_sin
    0.500000,    # hour_cos
    0.5,         # month_sin
    0.866025     # month_cos
]

# Prédiction
predicted_temp = predict_temperature(best_model, sample_features)
print(f"Température prédite: {predicted_temp:.2f} °C")
```

```python
from skopt import gp_minimize
from skopt import BayesSearchCV
from skopt.space import Real, Categorical
```

```python
spaces = {
    'adam': [
        Real(1e-5, 1e-2, name='learning_rate', prior='log-uniform'),
        Real(0.8, 0.99, name='beta_1'),
        Real(0.9, 0.999, name='beta_2')
    ],
    'sgd': [
        Real(1e-4, 1e-1, name='learning_rate'),
        Real(0.5, 0.99, name='momentum')
    ],
    'rmsprop': [
        Real(1e-5, 1e-2, name='learning_rate'),
        Real(0.8, 0.99, name='rho')
    ]
}
```

```python
def build_lstm_model(input_shape, optimizer='adam', learning_rate=0.001, momentum=0.0, beta_1=0.9, beta_2=0.999, rho=0.9):
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dropout, Dense
    from tensorflow.keras.optimizers import Adam, SGD, RMSprop

    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        LSTM(32),
        Dropout(0.2),
        Dense(1)
    ])

    # Créer l'optimiseur selon son nom
    if optimizer == 'adam':
        opt = Adam(learning_rate=learning_rate, beta_1=beta_1, beta_2=beta_2)
    elif optimizer == 'sgd':
        opt = SGD(learning_rate=learning_rate, momentum=momentum)
    elif optimizer == 'rmsprop':
        opt = RMSprop(learning_rate=learning_rate, rho=rho)
    else:
        raise ValueError(f"Optimiseur inconnu : {optimizer}")

    # Compiler le modèle
    model.compile(optimizer=opt, loss='mse', metrics=['mae'])
    return model
```

```python
def objective_function(params, optimizer_name, X_train, y_train, X_test, y_test):
    try:
        input_shape = X_train.shape[1:]

        # Appeler build_lstm_model avec les bons paramètres
        if optimizer_name == 'adam':
            lr, beta_1, beta_2 = params
            model = build_lstm_model(
                input_shape=input_shape,
                optimizer='adam',
                learning_rate=lr,
                beta_1=beta_1,
                beta_2=beta_2
            )

        elif optimizer_name == 'sgd':
            lr, momentum = params
            model = build_lstm_model(
                input_shape=input_shape,
                optimizer='sgd',
                learning_rate=lr,
                momentum=momentum
            )

        elif optimizer_name == 'rmsprop':
            lr, rho = params
            model = build_lstm_model(
                input_shape=input_shape,
                optimizer='rmsprop',
                learning_rate=lr,
                rho=rho
            )

        else:
            raise ValueError(f"Optimiseur inconnu : {optimizer_name}")

        # Entraîner le modèle
        history = model.fit(
            X_train, y_train,
            validation_data=(X_test, y_test),
            epochs=100,
            batch_size=32,
            callbacks=[EarlyStopping(patience=5, restore_best_weights=True, monitor='val_mae')],
            verbose=0
        )

        return min(history.history['val_mae'])

    except Exception as e:
        print(f"Erreur avec {params} : {str(e)}")
        return 1e10
```

```python
from skopt import gp_minimize
from skopt.space import Real

bo_results = {}

for opt_name in optimizers:
    print(f"\n=== Bayesian Optimization pour {opt_name.upper()} ===")

    try:
        res = gp_minimize(
            func=lambda params: objective_function(params, opt_name,
                                                 X_train, y_train,
                                                 X_test, y_test),
            dimensions=spaces[opt_name],
            n_calls=20,
            random_state=42,
            verbose=True
        )

        # ✅ This is the complete version with opt_result included:
        bo_results[opt_name] = {
            'best_params': res.x,
            'best_mae': res.fun,
            'optimizer': opt_name,
            'opt_result': res
        }

    except Exception as e:
        print(f"Optimization failed for {opt_name}: {str(e)}")
        bo_results[opt_name] = {
            'best_params': None,
            'best_mae': np.nan,
            'optimizer': opt_name
        }
```

```python
final_results = {}

for opt_name in bo_results:
    # Réentraînement avec les meilleurs paramètres
    best_params = bo_results[opt_name]['best_params']

    if opt_name == 'adam':
        optimizer_config = {'learning_rate': best_params[0], 'beta_1': best_params[1], 'beta_2': best_params[2]}
    elif opt_name == 'sgd':
        optimizer_config = {'learning_rate': best_params[0], 'momentum': best_params[1]}
    elif opt_name == 'rmsprop':
        optimizer_config = {'learning_rate': best_params[0], 'rho': best_params[1]}

    model = build_lstm_model(
        input_shape=input_shape,
        optimizer=opt_name,
        **optimizer_config
    )

    model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)

    final_results[opt_name] = {
        'test_mae': mae,
        'best_params': best_params
    }

# Comparaison avant/après optimisation
print("\nComparaison des performances:")
print("| Optimiseur | MAE Initial (test) | MAE Après BO (test) | Amélioration |")
print("|------------|--------------------|---------------------|--------------|")
for opt_name in optimizers:
    initial_mae = results[opt_name]['mae']
    optimized_mae = final_results[opt_name]['test_mae']
    improvement = (initial_mae - optimized_mae) / initial_mae * 100
    print(f"| {opt_name.upper()} | {initial_mae:.4f} | {optimized_mae:.4f} | {improvement:.1f}% |")
```
